```python
import matplotlib.pyplot as plt
```

```python
dims = [2**i for i in range(4, 12)]
x = [i for i in range(8)]

sequential = [0.013,
0.09,
1.019,
7.489,
46.866,
548.193,
5012.46,
78992.032]

inefficient = [0.8808,
0.9502,
0.9187,
0.8575,
1.1925,
1.9932,
4.3339,
13.3034]

naive = [0.6444,
0.7258,
0.6177,
0.7982,
1.319,
2.0804,
4.2897,
13.1451]

shared = [0.6287,
0.6389,
0.6242,
0.8191,
1.2154,
2.066,
4.3256,
13.2713]

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 20}

plt.rc('font', **font)
```

```python

title = "Matrix Multiplication Kernel Execution Time vs Matrix Dimensions"

fig = plt.figure(figsize=(15, 12))
axis = fig.add_axes([0,0,1,1])

axis.plot(x, sequential, linewidth=3)
axis.plot(x, inefficient)
axis.plot(x, naive)
axis.plot(x, shared)

plt.xticks(x, dims)
plt.legend(["Sequential", "Inefficient", "Naive", "Shared"])
axis.set_xlabel("Matrix Dimensions")
axis.set_ylabel("Execution Time (ms)")


axis.set_yscale('log')
plt.show()

extent = axis.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
fig.savefig("./{}.png".format(title), bbox_inches=extent.expanded(1.2, 1.2))
```
